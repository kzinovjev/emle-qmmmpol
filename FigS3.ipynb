```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('figure', facecolor='w')
```

```python
def read_auto(filename):
    return np.loadtxt(filename).T

def plot_auto_single(ax, filename):
    x, y = read_auto(filename)
    ax.plot(x, y, color='#1f77b4', lw=1.)
```

```python
filenames = [
    'wB97X/HSA_wB97X_ElEmb_La_auto.dat',
    'wB97X/HSA_MM_MM_La_auto.dat',
    'wB97X/HSA_ANI-2x_EMLE_La_auto.dat',
    'wB97X/HSA_wB97X_ElEmb_Lb_auto.dat',
    'wB97X/HSA_MM_MM_Lb_auto.dat',
    'wB97X/HSA_ANI-2x_EMLE_Lb_auto.dat',
]

systems = ['wB97X/MM', 'MM', 'ANI-2x(EMLE)/MM']
```

```python
fig, axs = plt.subplots(2, 3, figsize=(14, 4), sharex=True)
fig.subplots_adjust(wspace=0.25, hspace=0.2)

for i, (system, filename, ax) in enumerate(zip(systems * 2, filenames, axs.flatten())):
    plot_auto_single(ax, filename)
    ax.patch.set_alpha(0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xlim([0, 2000])
    ax.set_ylim([-0.01, 0.02])
    yticks = np.arange(-0.01, 0.021, 0.01)
    ax.set_yticks(yticks)
    for ytick in yticks:
        ax.axhline(ytick, linewidth=0.5, color='lightgrey', zorder=0)
    if i > 2:
        ax.set_xlabel('Time (fs)')
    else:
        ax.set_title(system)
    
fig.text(0.07, 0.5, 'wB97X autocorrelation (eV$^{2}$)', va='center', rotation='vertical', fontsize='large')
fig.savefig('FigS3.png', dpi=300, bbox_inches='tight')
plt.show()
```
